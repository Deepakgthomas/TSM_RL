```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pr_dqn_223 = pd.read_csv("dqn_stacked_Freewa_223_3.out", header=None, names = ["pr_dqn_223"])
pr_dqn_563 = pd.read_csv("dqn_stacked_Freewa_563_3.out", header=None, names = ["pr_dqn_563"])
pr_dqn_2004 = pd.read_csv("dqn_stacked_Freewa_2004_3.out", header=None, names = ["pr_dqn_2004"])
pr_dqn_7779 = pd.read_csv("dqn_stacked_Freewa_7779_3.out", header=None, names = ["pr_dqn_7779"])
pr_dqn_147 = pd.read_csv("dqn_stacked_Freewa_147_3.out", header=None, names = ["pr_dqn_147"])
```

```python
pr_dqn_merged = pr_dqn_223.join(pr_dqn_563)
pr_dqn_merged = pr_dqn_merged.join(pr_dqn_2004)
pr_dqn_merged = pr_dqn_merged.join(pr_dqn_7779)
pr_dqn_merged = pr_dqn_merged.join(pr_dqn_147)
pr_dqn_merged['mean'] = pr_dqn_merged.iloc[:, 0:5].mean(axis=1)
```

```python
pr_tsm_147 = pd.read_csv("tsm_Freewa_147_3_3.out", header=None, names = ["pr_tsm_147"])
pr_tsm_223 = pd.read_csv("tsm_Freewa_223_3_3.out", header=None, names = ["pr_tsm_223"])
pr_tsm_563 = pd.read_csv("tsm_Freewa_563_3_3.out", header=None, names = ["pr_tsm_563"])
pr_tsm_2004 = pd.read_csv("tsm_Freewa_2004_3_3.out", header=None, names = ["pr_tsm_2004"])
pr_tsm_7779 = pd.read_csv("tsm_Freewa_7779_3_3.out", header=None, names = ["pr_tsm_7779"])
```

```python
pr_tsm_merged = pr_tsm_147.join(pr_tsm_223)
pr_tsm_merged = pr_tsm_merged.join(pr_tsm_563)
pr_tsm_merged = pr_tsm_merged.join(pr_tsm_2004)
pr_tsm_merged = pr_tsm_merged.join(pr_tsm_7779)
pr_tsm_merged['mean'] = pr_tsm_merged.iloc[:, 0:5].mean(axis=1)
```

```python
pr_flare_147 = pd.read_csv("flare_Freewa_147_3.out", header=None, names = ["pr_flare_147"])
pr_flare_223 = pd.read_csv("flare_Freewa_223_3.out", header=None, names = ["pr_flare_223"])
pr_flare_563 = pd.read_csv("flare_Freewa_563_3.out", header=None, names = ["pr_flare_563"])
pr_flare_2004 = pd.read_csv("flare_Freewa_2004_3.out", header=None, names = ["pr_flare_2004"])
pr_flare_7779 = pd.read_csv("flare_Freewa_7779_3.out", header=None, names = ["pr_flare_7779"])
```

```python
pr_flare_merged = pr_flare_147.join(pr_flare_223)
pr_flare_merged = pr_flare_merged.join(pr_flare_563)
pr_flare_merged = pr_flare_merged.join(pr_flare_2004)
pr_flare_merged = pr_flare_merged.join(pr_flare_7779)
pr_flare_merged['mean'] = pr_flare_merged.iloc[:, 0:5].mean(axis=1)
```

```python
pd.set_option('display.max_rows', None)
```

```python
plt.scatter(np.arange(len(pr_flare_merged['mean'])), pr_flare_merged['mean'], label = "Average Flare for Freeway")

plt.scatter(np.arange(len(pr_tsm_merged['mean'])), pr_tsm_merged['mean'], label = "Average TSRL for Freeway")
plt.scatter(np.arange(len(pr_dqn_merged['mean'])), pr_dqn_merged['mean'], label = "Average DDQN-PER for Freeway")

plt.legend()
```

```python
pr_tsm_merged['mean'].sum(axis=0)
```

```python
pr_dqn_merged['mean'].sum(axis = 0)
```

```python
pr_flare_merged['mean'].sum(axis = 0)
```
